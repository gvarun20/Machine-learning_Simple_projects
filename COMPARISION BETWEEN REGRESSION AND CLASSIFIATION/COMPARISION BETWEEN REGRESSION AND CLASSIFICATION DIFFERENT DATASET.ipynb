# 1) REGRESSION FOR  STUDY PASS DATASET

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate a simple dataset
np.random.seed(42)
hours_studied = np.random.rand(20, 1) * 10
scores = 2 * hours_studied + 5 + np.random.randn(20, 1) * 2

# Train a linear regression model
regression_model = LinearRegression()
regression_model.fit(hours_studied, scores)

# Make predictions
predictions = regression_model.predict(hours_studied)

# Plot the regression line
plt.scatter(hours_studied, scores, label='Actual Data')
plt.plot(hours_studied, predictions, color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.title('Regression: Predicting Scores based on Hours Studied')
plt.legend()
plt.show()

# Evaluate the model
mse = mean_squared_error(scores, predictions)
print(f'Mean Squared Error: {mse}')
```

# 1.B)CLASSIFICATION FOR  STUDY PASS DATASET

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split

# Convert scores to binary labels for classification
pass_fail = (scores > np.median(scores)).astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(hours_studied, pass_fail, test_size=0.2, random_state=42)

# Train a logistic regression model
classification_model = LogisticRegression()
classification_model.fit(X_train, y_train)

# Make predictions
class_predictions = classification_model.predict(X_test)

# Plot the decision boundary
plt.scatter(hours_studied, pass_fail, label='Actual Data')
plt.plot(hours_studied, classification_model.predict(hours_studied), color='red', label='Decision Boundary')
plt.xlabel('Hours Studied')
plt.ylabel('Pass/Fail (1/0)')
plt.title('Classification: Predicting Pass/Fail based on Hours Studied')
plt.legend()
plt.show()

# Evaluate the model
accuracy = accuracy_score(y_test, class_predictions)
conf_matrix = confusion_matrix(y_test, class_predictions)
print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
```

# 2)REGRESSION FOR IRIS DATASET

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X_regression = iris.data[:, 0].reshape(-1, 1)  # Sepal length as input
y_regression = iris.data[:, 1]  # Sepal width as output

# Train a linear regression model
regression_model = LinearRegression()
regression_model.fit(X_regression, y_regression)

# Make predictions
predictions_regression = regression_model.predict(X_regression)

# Plot the regression line
plt.scatter(X_regression, y_regression, label='Actual Data')
plt.plot(X_regression, predictions_regression, color='red', label='Regression Line')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Regression: Predicting Sepal Width based on Sepal Length')
plt.legend()
plt.show()

# Evaluate the model
mse_regression = mean_squared_error(y_regression, predictions_regression)
print(f'Mean Squared Error: {mse_regression}')
```

# 2.B)CLASSIFICATION FOR IRIS DATASET

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Convert sepal width to binary labels for classification
y_classification = (iris.data[:, 1] > np.median(iris.data[:, 1])).astype(int)

# Split the data into training and testing sets
X_train_classification, X_test_classification, y_train_classification, y_test_classification = train_test_split(
    iris.data[:, :2], y_classification, test_size=0.2, random_state=42
)

# Train a logistic regression model
classification_model = LogisticRegression()
classification_model.fit(X_train_classification, y_train_classification)

# Make predictions
class_predictions_classification = classification_model.predict(X_test_classification)

# Plot the decision boundary
plt.scatter(X_test_classification[:, 0], X_test_classification[:, 1], c=y_test_classification, cmap=plt.cm.Paired, label='Actual Data')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Classification: Predicting High/Low Sepal Width based on Sepal Length')
plt.legend()
plt.show()

# Evaluate the model
accuracy_classification = accuracy_score(y_test_classification, class_predictions_classification)
conf_matrix_classification = confusion_matrix(y_test_classification, class_predictions_classification)
print(f'Accuracy: {accuracy_classification}')
print(f'Confusion Matrix:\n{conf_matrix_classification}')
```

# 3.B)CLASSIFICATION FOR Breast Cancer DATASET

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.datasets import load_breast_cancer

# Load the Breast Cancer dataset
breast_cancer = load_breast_cancer()

# Convert labels to binary: 0 for benign, 1 for malignant
y_classification = breast_cancer.target

# Split the data into training and testing sets
X_train_classification, X_test_classification, y_train_classification, y_test_classification = train_test_split(
    breast_cancer.data[:, :2], y_classification, test_size=0.2, random_state=42
)

# Train a logistic regression model
classification_model = LogisticRegression()
classification_model.fit(X_train_classification, y_train_classification)

# Make predictions
class_predictions_classification = classification_model.predict(X_test_classification)

# Create labels for actual data points
actual_labels = ['Has Cancer' if label == 1 else 'No Cancer' for label in y_test_classification]

# Plot the decision boundary
plt.scatter(X_test_classification[:, 0], X_test_classification[:, 1], c=y_test_classification, cmap=plt.cm.Paired, label='Predicted Data', edgecolors='k')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Classification: Predicting Cancer Status based on Mean Radius and Mean Texture')

# Plot decision boundary
h = .02  # step size in the mesh
x_min, x_max = X_test_classification[:, 0].min() - 1, X_test_classification[:, 0].max() + 1
y_min, y_max = X_test_classification[:, 1].min() - 1, X_test_classification[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = classification_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

# Highlight the actual data points with cancer labels
for i, txt in enumerate(actual_labels):
    plt.annotate(txt, (X_test_classification[i, 0], X_test_classification[i, 1]), ha='right', va='bottom', color='k', fontsize=8)

plt.legend()
plt.show()

# Evaluate the model
accuracy_classification = accuracy_score(y_test_classification, class_predictions_classification)
conf_matrix_classification = confusion_matrix(y_test_classification, class_predictions_classification)
print(f'Accuracy: {accuracy_classification}')
print(f'Confusion Matrix:\n{conf_matrix_classification}')
```
