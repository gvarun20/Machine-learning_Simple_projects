```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data)

# Reduce dimensionality for visualization
pca = PCA(n_components=2)
data_reduced = pca.fit_transform(data)

# Plot the true labels
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(data_reduced[:, 0], data_reduced[:, 1], c=target, cmap='viridis')
plt.title('True Labels')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Plot the clustering results
plt.subplot(1, 2, 2)
plt.scatter(data_reduced[:, 0], data_reduced[:, 1], c=clusters, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load the digits dataset
digits = load_digits()
data = digits.data
true_labels = digits.target

# Apply K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
clusters = kmeans.fit_predict(data)

# Reduce dimensionality for visualization
pca = PCA(n_components=2)
data_reduced = pca.fit_transform(data)

# Plot the true labels
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(data_reduced[:, 0], data_reduced[:, 1], c=true_labels, cmap='viridis')
plt.title('True Labels')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Plot the clustering results
plt.subplot(1, 2, 2)
plt.scatter(data_reduced[:, 0], data_reduced[:, 1], c=clusters, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D

# Load the digits dataset
digits = load_digits()
data = digits.data
true_labels = digits.target

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=3, random_state=42)
data_3d = tsne.fit_transform(data)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
clusters = kmeans.fit_predict(data)

# Plot the 3D scatter plot with true labels and clustering results
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

# True labels
scatter_true = ax.scatter(data_3d[:, 0], data_3d[:, 1], data_3d[:, 2], c=true_labels, cmap='viridis', marker='o', label='True Labels')

# Clustering results
scatter_clusters = ax.scatter(data_3d[:, 0], data_3d[:, 1], data_3d[:, 2], c=clusters, cmap='viridis', marker='^', label='K-Means Clustering')

ax.set_title('3D Visualization of Digits Dataset')
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')

# Add a legend
ax.legend()

plt.show()
```
