```python
# on peut aussi charger des datasets depuis la librarie
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data
```

```python
filter_titanic = titanic_data[['survived','pclass','sex','age','sibsp','parch','fare','embarked','deck']]
```

```python
filter_titanic 
```

```python
# nettoyons nos donnes
final_titanic = filter_titanic.dropna(axis=0, how='any')
```

```python
final_titanic
```

```python
# relation entre les colonnes de notre jeu de donnes 
sns.pairplot(final_titanic)
```

```python
# comme les donnes sont classes en gnral en catgorie donc on va les visualis avec catplot
sns.catplot(x='pclass', y='age', data=final_titanic, hue='sex') # repartition du genre dans chaque classe
```

```python
sns.boxplot(x='pclass', y='age', data=final_titanic, hue='sex') # repartition du genre dans chaque classe
```

```python
sns.jointplot(x='age', y='fare', data=final_titanic, kind='kde') # pour voir la distributuion 
#de deux features en meme temps
```

```python
sns.heatmap(final_titanic.corr()) # plus utilis pour voir la correlation entre futures
```
